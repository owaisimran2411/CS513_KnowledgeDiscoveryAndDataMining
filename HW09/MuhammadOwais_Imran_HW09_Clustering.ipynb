## **Library import**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering, KMeans
```

## **Loading CSV File**

```python
df = pd.read_csv("./MuhammadOwais_Imran_HW09_Cluster_Dataset.csv")
print(df.shape)
df = df.dropna() # drops NaN values
print(df[df.isna().any(axis=1)]) # shows NaN values in data frame if it exists
df.head()
```

## **Extracting Class Column and dropping it from data frame, and Scaling the Dataframe**

```python
target = df["diagnosis"]
df.drop(columns=["id", "diagnosis"], inplace=True)

scaler = MinMaxScaler()
# Fit and transform the data
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df.head()
```

```python
groups = AgglomerativeClustering(n_clusters=2, metric='euclidean',linkage='single')

clusters=groups.fit_predict(df)

df_cluster=pd.DataFrame({'Actual':target,'Cluster':clusters})
# Create a cross-tabulation
cross_tab = pd.crosstab(df_cluster['Actual'], df_cluster['Cluster'])

print(cross_tab)
```

## **KMeans Clustering**

```python
num_clusters = 2  # Iris dataset has three species
kmeans = KMeans(n_clusters=num_clusters, random_state=12)
kmeans.fit(df)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

df_cluster=pd.DataFrame({'Actual':target,'Cluster':labels})
# Create a cross-tabulation
cross_tab = pd.crosstab(df_cluster['Actual'], df_cluster['Cluster'])

print(cross_tab)
```
