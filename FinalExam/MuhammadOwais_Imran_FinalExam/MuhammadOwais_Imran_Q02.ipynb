```python
# library import

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering, KMeans
```

```python
# reading data from csv file

df = pd.read_csv('./MuhammadOwais_Imran_NYNJ_zipcode_population.csv')
print(f'Shape: {df.shape}')

df.head()
```

```python
# dropping the target coloum
target = df['STATE']
df.drop(columns=['STATE', 'STATEFIPS'], inplace=True)
df.shape
```

```python
# scaling the data using MinMax Scaler (enable if needed)
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
```

```python
# initialzing and implementing hierarchical clustering
groups = AgglomerativeClustering(n_clusters=3, linkage='single')
clusters = groups.fit_predict(df)
df_cluster = pd.DataFrame({'Actual': target, 'Cluster': clusters})
cross_tab = pd.crosstab(df_cluster['Actual'], df_cluster['Cluster'])

df['Cluster_Actual'] = target
df['Cluster_Predicted'] = clusters

print(cross_tab)

df
```
