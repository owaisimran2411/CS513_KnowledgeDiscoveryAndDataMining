## **Library import**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

## **Loading CSV File**

```python
df = pd.read_csv("./MuhammadOwais_Imran_HW07_SVM_Dataset.csv")
print(df.shape)
df = df.dropna() # drops NaN values
print(df[df.isna().any(axis=1)]) # shows NaN values in data frame if it exists
df.head()
```

## **Extracting Class Column and dropping it from data frame**

```python
target = df["diagnosis"]
df.drop(columns=["id", "diagnosis"], inplace=True)
df
```

## **Standardizing Dividing the data into train (70%) and test (30%) split**

```python
## using standard scaler instead of min max scaler
scaler = StandardScaler()

# Fit and transform the data
df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print(df.head())

X_train, X_test, Y_train, Y_test = train_test_split(
    df, target, test_size=0.3, random_state=4, shuffle=True

)
```

## **Initializing & Predicting Results C5.0 Decision Classifier**

```python
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
print(f"Model Accuracy: {accuracy_score(Y_test, Y_pred)}")
print(f"Confusion Matrix:\n {confusion_matrix(Y_test, Y_pred)}\n")
print(f"Classification Report:\n {classification_report(Y_test, Y_pred)}\n")
```
